```python
# Sales Data Analysis - Brainwave Matrix Solutions
# Author: Gaurav
# Date: August 2025

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('default')
sns.set_palette("husl")

print("=" * 60)
print("SALES DATA ANALYSIS - BRAINWAVE MATRIX SOLUTIONS")
print("=" * 60)
print("üìä Comprehensive Sales Analysis Report")
print("üéØ Internship Task #1")
print("=" * 60)
```

```python
# 1. DATA LOADING AND OVERVIEW
print("\nüîç 1. DATA LOADING AND OVERVIEW")
print("-" * 40)

# Load the dataset
df = pd.read_csv('../data/sales_data.csv')

# Basic information
print(f"Dataset Shape: {df.shape}")
print(f"Date Range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Total Sales: ${df['Total_Sales'].sum():,.2f}")
print(f"Total Profit: ${df['Profit'].sum():,.2f}")
print(f"Unique Customers: {df['Customer_ID'].nunique()}")
print(f"Unique Products: {df['Product'].nunique()}")

# Display first few records
print("\nüìã Sample Data:")
df.head()
```

```python
# Check data types and missing values
print("üîç Data Info:")
df.info()
print("\nüîç Missing Values:")
print(df.isnull().sum())
```

```python
# 2. DATA PREPROCESSING
print("\nüõ†Ô∏è 2. DATA PREPROCESSING")
print("-" * 40)

# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract date components
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
df['Quarter'] = df['Date'].dt.quarter

# Calculate profit margin
df['Profit_Margin'] = (df['Profit'] / df['Total_Sales']) * 100

print("‚úÖ Data preprocessing completed!")
print("üìÖ Added date components: Year, Month, Day, Weekday, Quarter")
print("üí∞ Added profit margin calculation")

# Show updated dataframe info
df.head()
```

```python
# 3. KEY PERFORMANCE METRICS
print("\nüìä 3. KEY PERFORMANCE METRICS")
print("-" * 40)

# Basic statistics
print("üìà Key Performance Metrics:")
print(f"Average Order Value: ${df['Total_Sales'].mean():.2f}")
print(f"Average Profit Margin: {df['Profit_Margin'].mean():.2f}%")
print(f"Average Items per Order: {df['Quantity'].mean():.1f}")
print(f"Total Orders: {len(df):,}")

# Quick summary statistics
df.describe().round(2)
```

```python
# 4. SALES TREND ANALYSIS
print("\nüìà 4. SALES TREND ANALYSIS")
print("-" * 40)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Monthly Sales Trend
monthly_trend = df.groupby(df['Date'].dt.to_period('M'))['Total_Sales'].sum()
monthly_trend.plot(kind='line', ax=axes[0,0], color='blue', linewidth=2)
axes[0,0].set_title('Monthly Sales Trend', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Sales ($)')
axes[0,0].grid(True, alpha=0.3)

# 2. Category Performance
category_sales = df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', ax=axes[0,1], color='green')
axes[0,1].set_title('Sales by Category', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Sales ($)')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Regional Performance
region_sales = df.groupby('Region')['Total_Sales'].sum()
region_sales.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%')
axes[1,0].set_title('Sales Distribution by Region', fontsize=14, fontweight='bold')
axes[1,0].set_ylabel('')

# 4. Sales Channel Analysis
channel_sales = df.groupby('Sales_Channel')['Total_Sales'].sum()
channel_sales.plot(kind='bar', ax=axes[1,1], color='orange')
axes[1,1].set_title('Sales by Channel', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Sales Channel')
axes[1,1].set_ylabel('Sales ($)')
axes[1,1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('../visualizations/sales_overview.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 5. PRODUCT ANALYSIS
print("\nüõçÔ∏è 5. PRODUCT ANALYSIS")
print("-" * 40)

# Top products by sales
top_products = df.groupby('Product').agg({
    'Total_Sales': 'sum',
    'Quantity': 'sum',
    'Order_ID': 'count'
}).round(2).sort_values('Total_Sales', ascending=False)

print("üèÜ Top 10 Products by Sales:")
print(top_products.head(10))

# Product performance visualization
plt.figure(figsize=(12, 8))
top_10_products = top_products.head(10)
plt.barh(range(len(top_10_products)), top_10_products['Total_Sales'])
plt.yticks(range(len(top_10_products)), top_10_products.index)
plt.xlabel('Total Sales ($)')
plt.title('Top 10 Products by Sales Performance', fontsize=16, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('../visualizations/top_products.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 6. CUSTOMER ANALYSIS
print("\nüë• 6. CUSTOMER ANALYSIS")
print("-" * 40)

# Customer segment analysis
customer_analysis = df.groupby('Customer_Segment').agg({
    'Total_Sales': 'sum',
    'Profit': 'sum',
    'Order_ID': 'count',
    'Customer_ID': 'nunique'
}).round(2)

print("üë• Customer Segment Performance:")
print(customer_analysis)

# Customer behavior analysis
customer_behavior = df.groupby('Customer_ID').agg({
    'Total_Sales': 'sum',
    'Order_ID': 'count',
    'Quantity': 'sum'
}).round(2)

customer_behavior['Avg_Order_Value'] = customer_behavior['Total_Sales'] / customer_behavior['Order_ID']

print("\nüìä Customer Behavior Insights:")
print(f"Average Customer Lifetime Value: ${customer_behavior['Total_Sales'].mean():.2f}")
print(f"Average Orders per Customer: {customer_behavior['Order_ID'].mean():.1f}")
print(f"Average Order Value: ${customer_behavior['Avg_Order_Value'].mean():.2f}")

# Visualize customer segments
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
customer_analysis['Total_Sales'].plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Sales by Customer Segment', fontweight='bold')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)

plt.subplot(1, 2, 2)
customer_analysis['Customer_ID'].plot(kind='pie', autopct='%1.1f%%', colors=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Customer Distribution by Segment', fontweight='bold')
plt.ylabel('')

plt.tight_layout()
plt.savefig('../visualizations/customer_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 7. SEASONAL & TIME ANALYSIS
print("\nüóìÔ∏è 7. SEASONAL & TIME ANALYSIS")
print("-" * 40)

# Quarterly analysis
quarterly_sales = df.groupby(['Year', 'Quarter']).agg({
    'Total_Sales': 'sum',
    'Order_ID': 'count'
}).round(2)

print("üóìÔ∏è Quarterly Sales Performance:")
print(quarterly_sales)

# Day of week analysis
weekday_sales = df.groupby('Weekday')['Total_Sales'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

plt.figure(figsize=(15, 5))

# Weekday sales
plt.subplot(1, 3, 1)
weekday_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Day of Week', fontweight='bold')
plt.xlabel('Day of Week')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)

# Monthly sales
plt.subplot(1, 3, 2)
monthly_sales = df.groupby('Month')['Total_Sales'].sum()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.bar(range(1, 13), monthly_sales.values, color='lightgreen')
plt.xticks(range(1, 13), month_names, rotation=45)
plt.title('Sales by Month', fontweight='bold')
plt.ylabel('Total Sales ($)')

# Quarterly sales
plt.subplot(1, 3, 3)
quarter_sales = df.groupby('Quarter')['Total_Sales'].sum()
plt.bar(['Q1', 'Q2', 'Q3', 'Q4'], quarter_sales.values, color='coral')
plt.title('Sales by Quarter', fontweight='bold')
plt.ylabel('Total Sales ($)')

plt.tight_layout()
plt.savefig('../visualizations/time_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 8. PROFIT ANALYSIS
print("\nüí∞ 8. PROFIT ANALYSIS")
print("-" * 40)

# Profit margin by category
profit_by_category = df.groupby('Category').agg({
    'Profit': 'sum',
    'Total_Sales': 'sum',
    'Profit_Margin': 'mean'
}).round(2)

print("üí∞ Profitability by Category:")
print(profit_by_category)

# Profit margin by region
profit_by_region = df.groupby('Region').agg({
    'Profit': 'sum',
    'Profit_Margin': 'mean'
}).round(2)

print("\nüåç Profitability by Region:")
print(profit_by_region)

# Visualize profit analysis
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
profit_by_category['Profit'].plot(kind='bar', color='gold')
plt.title('Total Profit by Category', fontweight='bold')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
profit_by_category['Profit_Margin'].plot(kind='bar', color='lightcoral')
plt.title('Avg Profit Margin by Category', fontweight='bold')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
profit_by_region['Profit'].plot(kind='bar', color='lightblue')
plt.title('Total Profit by Region', fontweight='bold')
plt.ylabel('Profit ($)')
plt.xticks(rotation=0)

plt.subplot(2, 2, 4)
# Scatter plot: Sales vs Profit
plt.scatter(df['Total_Sales'], df['Profit'], alpha=0.6, color='purple')
plt.xlabel('Total Sales ($)')
plt.ylabel('Profit ($)')
plt.title('Sales vs Profit Relationship', fontweight='bold')

plt.tight_layout()
plt.savefig('../visualizations/profit_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 9. KEY FINDINGS & BUSINESS RECOMMENDATIONS

## üîë Key Findings
Based on the analysis, we've uncovered several critical insights:

1.  **Sales Growth:** There is a clear positive trend in monthly sales, indicating healthy business growth.
2.  **Top Performers:**
    * **Category:** `Electronics & Gadgets` is the highest revenue-generating category by a significant margin.
    * **Region:** The `North` region is the most profitable and contributes the largest portion of sales.
    * **Channel:** The `Online` channel outperforms `In-Store` and `Mobile App` sales.
3.  **Customer Behavior:**
    * The `Premium` customer segment, while smaller, has a high average order value.
    * The `Regular` segment represents the largest customer base.
4.  **Seasonal Trends:**
    * Sales consistently peak during **Q4**, likely due to holiday shopping.
    * Sales are generally higher on **weekends** (Friday, Saturday, Sunday).
5.  **Profitability Insights:**
    * While `Electronics & Gadgets` brings in the most sales, the **`Home & Kitchen`** category has a higher average profit margin.
    * The relationship between Sales and Profit is strong and positive.

## üöÄ Business Recommendations

1.  **Inventory & Marketing Focus:**
    * **Action:** Increase stock levels for `Electronics & Gadgets` and `Home & Kitchen` products, especially leading into **Q4**.
    * **Reasoning:** Capitalize on the highest-selling and highest-margin categories during the peak sales season to maximize revenue and profit.

2.  **Targeted Regional Campaigns:**
    * **Action:** Launch targeted digital marketing campaigns for the `North` region, focusing on promoting products through the `Online` channel.
    * **Reasoning:** Double down on the most successful region and channel combination to drive further growth.

3.  **Customer Segmentation Strategy:**
    * **Action:** Develop a loyalty program to encourage `Regular` customers to increase their spending and transition towards the `Premium` segment. Offer exclusive online deals to the `Premium` segment.
    * **Reasoning:** Increase customer lifetime value by nurturing the largest customer segment and rewarding the most valuable one.

4.  **Promote High-Margin Products:**
    * **Action:** Create special promotions or bundles featuring `Home & Kitchen` products to increase their sales volume.
    * **Reasoning:** Boosting sales of high-margin products will directly improve the company's overall profitability.
